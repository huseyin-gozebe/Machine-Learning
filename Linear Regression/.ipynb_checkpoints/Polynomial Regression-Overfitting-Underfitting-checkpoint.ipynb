## Evaluation on the Test Set

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = model_poly.predict(X_test)
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance:")
    print("--------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)
```

```python
eval_metric(y_test, y_pred)
```

```python
y_train_pred = model_poly.predict(X_train)
```

```python
eval_metric(y_train, y_train_pred)
```

## Let's find optimal degree of poly

```python
# With "poly" function below we calculate everything that we did above
```

```python
def poly(d):
    
    train_rmse_errors = []
    test_rmse_errors = []
    
    for i in range(1, d):
        polynomial_converter = PolynomialFeatures(degree = i, include_bias =False)
        poly_features = polynomial_converter.fit_transform(X)
        
        X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
        
        model = LinearRegression(fit_intercept=True)
        model.fit(X_train, y_train)
        
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)
        
        train_RMSE = np.sqrt(mean_squared_error(y_train,train_pred))
        test_RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
        
        train_rmse_errors.append(train_RMSE)
        test_rmse_errors.append(test_RMSE)
        
    return pd.DataFrame({"train_rmse_errors": train_rmse_errors, "test_rmse_errors":test_rmse_errors}, \
                        index=[i for i in range(1,d)])
        
```

```python
poly(10)
```

```python
# If there is a huge difference between train and test scores that informs us on overfitting.
```

```python
# degree equals 5 is the start of overfitting area. 
```

```python
plt.plot(range(1,10), poly(10)["train_rmse_errors"], label = "TRAIN") 
plt.plot(range(1,10), poly(10)["test_rmse_errors"], label = "TEST") 
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
```

```python
# To better understand and check overfitting we make the degree 6.
```

```python
plt.plot(range(1,6), poly(6)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,6), poly(6)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
```

```python
# 1 and 2 underfitting area
# 2 thorugh 4 ideal area
# 4 - ~ overfitting area
```

```python
# To select 4 as degree is kind of dangerous since it is very close to overfitting area. 
# So, we select 3 as optimal degree
```

## Overfitting

```python
over_poly_converter = PolynomialFeatures(degree = 5, include_bias=False)
```

```python
over_model = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(over_poly_converter.fit_transform(X), y, \
                                                    test_size=0.3, random_state=101)
```

```python
over_model.fit(X_train, y_train)
```

```python
y_pred_over = over_model.predict(X_test)
```

```python
eval_metric(y_test, y_pred_over)
```

```python
y_train_over = over_model.predict(X_train)
```

```python
eval_metric(y_train, y_train_over)
```
