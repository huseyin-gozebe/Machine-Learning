# Regularization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop("sales", axis =1)
y = df["sales"]
```

### Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures

polynomial_converter = PolynomialFeatures(degree=5, include_bias=False) 

poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

```python
poly_features
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)  
```

## Scaling the Data

```python
# If we don't scale our data, a feature with large values will have larger coefficients 
# that directs us to misevaluation about importance of features.
# If you are unsure about scaling data do scale them.
# Tree based models do nor require scaling.
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

# RobustScaler is used if there are outliers in data set

scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
# We are scaling only training data, not all data to avoid from data leakage
```

```python
X_train = scaler.transform(X_train)

# Z-score scaling (also Z-score normalization) with mean is 0 and sdandart deviation is 1. 
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
pd.DataFrame(X_train).agg(["mean", "std"]).round()
```

```python
pd.DataFrame(X_test).agg(["mean", "std"]).round()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score

def eval_metrics(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance:")
    print("--------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
y_train_pred = lm.predict(X_train)
```

```python
eval_metrics(y_train, y_train_pred)
```

```python
eval_metrics(y_test, y_pred)
```

```python
lm_scores = {"lm_train": {"rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred)),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),
    "R2" : r2_score(y_train, y_train_pred)}, 

    "lm_test": {"rmse" : np.sqrt(mean_squared_error(y_test, y_pred)),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "R2" : r2_score(y_test, y_pred)}}
ls =pd.DataFrame(lm_scores)
ls
```

## Multicollinearity

```python
def color_red(val):
    
    if val > 0.90 and val < 0.99:
        color = 'red'
    else:
        color = 'black'
    return f'color: {color}'

# %90 ve zeri bir multicorolasyon gstergesidir.
```

## Grid Search

```python
ridge ve lasso sadece alpha y kullanr. (alpha ya ilave olarak)  l1_ratio yu elastic te kullanyoruz. 

l1_ratio deeri ile lassaya ne oranda benzediini anlyoruz; 1 ise %100 lasso oluyor.
```

```python
her algorithmann lasso ve ridge gibi kendi class  iinde hemgrid search hemde CV yapan(RidgeCV,LassoCV gibi) 
bir modeli olmayabilir. 
ite bu durumda bizim yerimize modele ikisini birden uygulayan methoda GridsearchCV diyoruz
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
elastic_model = ElasticNet(max_iter=10000)
```

```python
param_grid ={"alpha":[0.01, 0.02, 0.1, 1, 5, 10, 50, 100],
            "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]}

# kullanabileceimiz deerler az ok belli. o aral kullanyoruz.
# l1_ratio zaten 0 la 1 arasnda oluyor. 
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =10, verbose =2)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```
